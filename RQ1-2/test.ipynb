```python
import pandas as pd

# 讀取 CSV 檔案
# df = pd.read_csv('./articles.csv', encoding='utf-8')
df = pd.read_csv('./articles.csv', encoding='ISO-8859-1')

# 只取 content 欄位
contents = df['content'].tolist()

# 取前 4 篇文章
articles = {f'Article_{i+1}': content for i, content in enumerate(contents[:4])}

# 取代 prompt 內的佔位符
prompt_with_articles = {title: prompt + content for title, content in articles.items()}

print(len(prompt_with_articles))  # 應該會輸出 4
```

```python
responses = {filename: get_response(prompt) for filename, prompt in prompt_with_articles.items()}
```

```python
import pandas as pd
from io import StringIO

for filename, response in responses.items():
    content = response[response.find("*****\n")+5:response.rfind("\n*****")] #格式可以改成你怎麼切分
    # print(content)
    df = pd.read_csv(StringIO(content), sep="/", header=None, names=["Entity", "Entity Type", "Reaction or Emotion", "Reason"], on_bad_lines='skip')
    # df = pd.read_csv(StringIO(content), sep="/")
    df.to_csv(f'./responses/{filename}.csv', index=False)
    print(f'{filename} done')
```
